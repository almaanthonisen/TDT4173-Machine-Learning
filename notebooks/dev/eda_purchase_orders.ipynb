```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

po = pd.read_csv("../../data/kernel/purchase_orders.csv", low_memory=False)

# Parse dates
po["delivery_date"] = pd.to_datetime(po["delivery_date"], errors="coerce", utc=True)
po["created_date_time"] = pd.to_datetime(po["created_date_time"], errors="coerce", utc=True)
po["modified_date_time"] = pd.to_datetime(po["modified_date_time"], errors="coerce", utc=True)

print("Shape:", po.shape)
print("Columns:", po.columns.tolist())
print("Date range:", po["delivery_date"].min(), "→", po["delivery_date"].max())

po.head()
```
